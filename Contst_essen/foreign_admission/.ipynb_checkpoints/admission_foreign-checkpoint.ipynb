# FILLING UP NULL VALUES

```python
# Undergraduate_University                                                         2
# Undergraduate_CGPA                                                               7
# GRE                                                                             30
# IELTS_TOEFL                                                                      9
# Subject_got_chance                                                               7
# Research                                                                         2
# If you're interested, you may add your work position before your admission?    125
```

```python
df["If you're interested, you may add your work position before your admission?"].dtype
```

```python
df.rename(columns={"If you're interested, you may add your work position before your admission?": "work_position_before_admission"},inplace= True)
```

## Object fill

```python
df['Undergraduate_University']=df['Undergraduate_University'].fillna(df['Undergraduate_University'].mode()[0])
# df['Subject_got_chance']=df['Subject_got_chance'].fillna(df['Subject_got_chance'].mode()[0])
# df["If you're interested, you may add your work position before your admission?"]=df["If you're interested, you may add your work position before your admission?"].fillna(df["If you're interested, you may add your work position before your admission?"].mode()[0])
```

## Numeric

```python
df['Undergraduate_CGPA'] = df['Undergraduate_CGPA'].fillna(df['Undergraduate_CGPA'].mean())
df['GRE'] = df['GRE'].fillna(df['GRE'].mean())
df['IELTS_TOEFL'] = df['IELTS_TOEFL'].fillna(df['IELTS_TOEFL'].mean())
df['Research'] = df['Research'].fillna(df['Research'].mean())
```
